```python
query = { "message": {"query_graph": {
  "edges": {
    "e0": {
      "object": "n1",
      "predicates": [
        "biolink:gene_associated_with_condition",
        "biolink:condition_associated_with_gene",
        "biolink:genetic_association"
      ],
      "subject": "n0"
    }
  },
  "nodes": {
    "n0": {
      "categories": [
        "biolink:Gene"
      ],
      "ids": [
        "NCBIGene:727897"
      ]
    },
    "n1": {
      "categories": [
        "biolink:Disease"
      ]
    }
  }
}}}
```

```python
r = requests.post("https://strider.renci.org/1.3/query",json=query)
```

```python
message = r.json()["message"]
for edge_id, edge in message["knowledge_graph"]["edges"].items():
    print(edge["subject"],edge["object"], edge["predicate"])
```

```python
from collections import defaultdict
counts = defaultdict(int)
for result in message["results"]:
    n0s = frozenset([node["id"] for node in result["node_bindings"]["n0"]])
    n1s = frozenset([node["id"] for node in result["node_bindings"]["n1"]])
    nodes = (n0s, n1s)
    counts[nodes]+=1
for nodekey, num in counts.items():
    if num > 1:
        print(nodekey, num)
```
