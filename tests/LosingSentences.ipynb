```python
query = {
    "message":{
        "query_graph":{
            "nodes": {
                "s": {
                    "categories": ["biolink:ChemicalEntity"],
                    "ids": ["CHEMBL.COMPOUND:CHEMBL509176"]
                },
                "o": {
                    "categories": ["biolink:Disease"]
                }
            },
            "edges": {
                "e": {
                    "subject": "s",
                    "object": "o",
                    "predicates": ["biolink:treats"]
                }
            }
        }
    }
}
```

```python
r = requests.post("https://arax.ncats.io/api/rtxkg2/v1.3/query",json=query)
```
