```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("comparison.tsv",sep="\t")
```

```python
len(df)
```

```python
df.columns
```

```python
plt.plot(df["UMLS IC"], df["Ubergraph IC"], ".")
plt.xlabel("UMLS")
plt.ylabel("Ubergraph")
plt.show()
```

```python
groups = df.groupby(by="BiolinkType")
```

```python
len(groups)
```

```python
for group in groups:
    name = group[0]
    data = group[1]
    plt.plot(data["UMLS IC"], data["Ubergraph IC"], ".")
    plt.title(name)
    plt.xlabel("UMLS IC")
    plt.ylabel("Ubergraph IC")
    plt.show()
```

```python
df["diff"] = df["Ubergraph IC"] - df["UMLS IC"]
```

```python
dg = df.sort_values(by="diff", ascending=False)
```

```python
sorted_groups = dg.groupby("BiolinkType")
for groupname, group in sorted_groups:
    display(group.head(10))
```
