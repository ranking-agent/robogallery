```python
with open('Ferguson_GeneTranscripts_with_ids_2.csv','r') as inf:
    h = inf.readline()
    ids = []
    for line in inf:
        x = line.strip().split(',')
        ids.append((x[0],x[1]))
```

```python
def make_query(in_ids,n=0):
    template_query = {
        "message":{"query_graph":{
      "edges": {

      },
      "nodes": {
        "g0": {
          "categories": [
            "biolink:Gene"
          ]
        }
      }},
      "knowledge_graph":
        {"nodes":
         {

         }
        ,
        "edges":
         {},

      },
                   "results":[
        ]
                  },
      "workflow": [{"id":"enrich_results"}]
                  }

    if n == 0:
        n = len(in_ids)
    for name,nodeid in in_ids[:n]:
        template_query['message']['knowledge_graph']['nodes'][nodeid] = {"categories": [
    "biolink:Entity",
    "biolink:GeneOrGeneProduct",
    "biolink:MacromolecularMachineMixin",
    "biolink:BiologicalEntity",
    "biolink:Gene",
    "biolink:NamedThing"
    ]}
        template_query['message']['results'].append({"node_bindings": {"g0": [{"id":nodeid}]}, "edge_bindings":{}},)
    
    return template_query
```

```python
def print_result(epk):
    for r in epk['message']['results']:
        numgenes = len(r['node_bindings']['g0'])

        if numgenes == 1:
            continue
        keys = r['node_bindings'].keys()
        for k in keys:
            if k.startswith('_'):
                extra_node_id = r['node_bindings'][k][0]['id']
                extra_node_name = epk['message']['knowledge_graph']['nodes'][extra_node_id]['name']
                print(numgenes, extra_node_id, extra_node_name)
```

```python
query = make_query(ids)
r = aragorn(query)
print_result(r)
```

```python
query = make_query(ids,n=100)
r = aragorn(query)
print_result(r)
```
